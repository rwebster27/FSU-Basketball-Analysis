```python
import pandas as pd
import numpy as np
```

```python
GameStats1 = pd.read_csv('GameStats1.csv')
GameStats2 = pd.read_csv('GameStats2.csv')
GameStats3 = pd.read_csv('GameStats3.csv')
GameStats4 = pd.read_csv('GameStats4.csv')
```

## Dropping Uneeded Columns

```python
GameStats1 = GameStats1.drop([31])
```

```python
GameStats2 = GameStats2.drop(columns = ['Date', 'Opponent', 'Result','PTS', 'PTS/POSS'], axis = 1)
GameStats2 = GameStats2.drop([31])
```

```python
GameStats3 = GameStats3.drop(columns = ['Date', 'Opponent', 'Result','OFF REB%', 'DEF REB%', 'REB%'])
GameStats3 = GameStats3.drop([31])
```

```python
GameStats4 = GameStats4.drop(columns = ['Date', 'Opponent', 'Result','JUMPSHOT','DUNK','LAYUP','TIPIN','HOOK','PF.1','TECH'])
GameStats4 = GameStats4.drop([31])
```

## Adding all Data to One DataFrame

```python
GameStats_All = pd.concat([GameStats1, GameStats2, GameStats3, GameStats4], axis = 1)
```

```python
GameStats_All
```

## Clean Opponent

```python
GameStats_All['Home/Away'] = GameStats_All['Opponent'].str[0]
```

```python
GameStats_All['Opponent'] = GameStats_All['Opponent'].str[1:]
```

```python
GameStats_All.loc[(GameStats_All['Opponent'].str[0] == "s"), 'Opponent'] = GameStats_All['Opponent'].str[1:]
```

```python
GameStats_All.loc[(GameStats_All['Home/Away'] == "@"),'Home/Away'] = 'Away'
GameStats_All.loc[(GameStats_All['Home/Away'] == "v"),'Home/Away'] = 'Home'
```

## Clean Result

```python
GameStats_All['Win/Loss'] = GameStats_All['Result'].str[0]
```

```python
GameStats_All['Result'] = GameStats_All['Result'].str[2:]
```

```python
GameStats_All['PTs/A'] = GameStats_All['Result'].str.split("-").str[1]
```

```python
GameStats_All['PTs/A'] = GameStats_All['PTs/A'].astype('float64')
```

```python
GameStats_All
```

## Cleaning Team Name

```python
Names = {'Full Name': ['Pittsburgh','Florida','W Carolina','Chattanooga','St Fran (PA)','Chicago St','Tennessee','Purdue','Indiana','Clemson','N Florida','S Florida','North Alabama','GA Tech','Louisville','Wake Forest','Virginia','Miami (FL)','Notre Dame','VA Tech','N Carolina','Duke','Syracuse','NC State','Boston Col']}
```

```python
Team_Names = pd.DataFrame(data = Names)
```

```python
Team_Names['Abv'] = GameStats_All['Opponent']
```

```python
Team_Names['Abv'][19] = 'VT'
Team_Names['Abv'][20] = 'UNC'
Team_Names['Abv'][21] = 'DUKE'
Team_Names['Abv'][22] = 'SYR'
Team_Names['Abv'][23] = 'NCST'
Team_Names['Abv'][24] = 'BC'
```

```python
Team_Names = Team_Names[['Abv','Full Name']]
```

```python
GameStats_All['Opponent'] = GameStats_All['Opponent'].str.lstrip()
Team_Names['Abv'] = Team_Names['Abv'].str.lstrip()
```

```python
Names = Team_Names.set_index('Abv').T.to_dict('index')
```

```python
GameStats_All['Opponent'] = GameStats_All['Opponent'].map(Names['Full Name'])
```
